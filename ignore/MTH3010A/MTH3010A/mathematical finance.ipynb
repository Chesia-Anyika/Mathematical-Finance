# Capital Asset Pricing Model

```python
import numpy as np
import pandas as pd
import yfinance as yf

def calculate_expected_return(risk_free_rate, beta, market_return):
    """
    Calculate the expected return of an asset using the Capital Asset Pricing Model (CAPM).

    Parameters:
    risk_free_rate (float): The risk-free rate.
    beta (float): The beta of the asset.
    market_return (float): The expected return of the market.

    Returns:
    float: The expected return of the asset.
    """
    return risk_free_rate + beta * (market_return - risk_free_rate)

# Example inputs
risk_free_rate = 0.03  # 3%
beta = 1.2
market_return = 0.08  # 8%

expected_return = calculate_expected_return(risk_free_rate, beta, market_return)
print(f"Expected Return: {expected_return:.2%}")

# Optional: Calculating Beta using historical stock and market data
def calculate_beta(asset_returns, market_returns):
    """
    Calculate the beta of an asset based on historical returns.

    Parameters:
    asset_returns (pd.Series): Historical returns of the asset.
    market_returns (pd.Series): Historical returns of the market.

    Returns:
    float: The beta of the asset.
    """
    covariance = np.cov(asset_returns, market_returns)[0, 1]
    market_variance = np.var(market_returns)
    beta = covariance / market_variance
    return beta

# Example: Fetch historical data for a stock and a market index
asset_ticker = 'AAPL'
market_ticker = '^GSPC'  # S&P 500

asset_data = yf.download(asset_ticker, start="2020-01-01", end="2023-01-01")
market_data = yf.download(market_ticker, start="2020-01-01", end="2023-01-01")

# Calculate daily returns
asset_returns = asset_data['Adj Close'].pct_change().dropna()
market_returns = market_data['Adj Close'].pct_change().dropna()

# Calculate Beta
beta = calculate_beta(asset_returns, market_returns)
print(f"Calculated Beta: {beta:.2f}")

# Calculate Expected Return using the calculated Beta
expected_return_calculated_beta = calculate_expected_return(risk_free_rate, beta, market_return)
print(f"Expected Return with Calculated Beta: {expected_return_calculated_beta:.2%}")
```
