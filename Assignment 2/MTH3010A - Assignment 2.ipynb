Consider the initial value problem (IVP) given by the following differential equation:
$$\frac{dy}{dt} = f(t,y) = t - y$$

## Exact Solution
### Step 1: Compute the exact solution
First we rewrite the differential equation into a first order linear differential equation

$$\frac{dy}{dt} + y = t$$

Then we identify the integrating factor: 

$$\mu(t) = e^{\int 1dt} = e^t$$

Multiply both sides by the integrating factor: 

$$e^t \frac{dy}{dt} + e^ty = te^t$$

The left hand side of the equation is the derivative of $ye^t$:

$$\frac{d}{dt}(ye^t) = te^t$$

Integrate both sides with respect to t:

$$\int \frac{d}{dt}(ye^t) = \int te^t\ dt\\ 
ye^t = \int te^t\ dt$$

Solve the right hand side integral using integration by parts:

$$\text{Let } u = t\ \text{and } dv = e^tdt \\
\text{Then }du = dt\ \text{and }v = e^t\\
\text{Integration by parts formula: }\int u\ sv = uv-\int v\ du\\
\text{Applying the formula: } \int te^tdt = te^t - \int e^tdt = te^t-e^t$$

Combine the results of the integration of the left and right hand sides:

$$ye^t = te^t - e^t +C\\
ye^t = e^t(t-1) + C$$

Solve for $y$:
$$y=t-1+Ce^{-t}$$

Use the initial condition $y(0) = 1$ to find C:
$$1=0-1+Ce^0\\
1=-1+C\\
C=2$$

Thus the exact solution is:
$$y(t) = t-1 +2e^{-t}$$

**Write a python code to check your results and plot all the numerical solutions (the
three) along with the exact solution in one graph.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential equation
def f(t, y):
    return t - y

# Euler Method
def euler_method(h, N):
    t = 0
    y = 1
    t_values = [t]
    y_values = [y]
    for i in range(N):
        y = y + h * f(t, y)
        t = t + h
        t_values.append(t)
        y_values.append(y)
    return t_values, y_values

# Heun Method
def heun_method(h, N):
    t = 0
    y = 1
    t_values = [t]
    y_values = [y]
    for i in range(N):
        k1 = f(t, y)
        k2 = f(t + h, y + h * k1)
        y = y + (h / 2) * (k1 + k2)
        t = t + h
        t_values.append(t)
        y_values.append(y)
    return t_values, y_values

# Runge-Kutta 4th Order Method
def runge_kutta_method(h, N):
    t = 0
    y = 1
    t_values = [t]
    y_values = [y]
    for i in range(N):
        k1 = f(t, y)
        k2 = f(t + h / 2, y + h * k1 / 2)
        k3 = f(t + h / 2, y + h * k2 / 2)
        k4 = f(t + h, y + h * k3)
        y = y + (h / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
        t = t + h
        t_values.append(t)
        y_values.append(y)
    return t_values, y_values

# Exact Solution
def exact_solution(t):
    return t - 1 + 2 * np.exp(-t)

# Parameters
h = 0.1
N = int(2 / h)

# Compute numerical solutions
t_euler, y_euler = euler_method(h, N)
t_heun, y_heun = heun_method(h, N)
t_rk, y_rk = runge_kutta_method(h, N)
t_exact = np.linspace(0, 2, 100)
y_exact = exact_solution(t_exact)

# Plot the solutions
plt.plot(t_euler, y_euler, 'r', label='Euler Method')
plt.plot(t_heun, y_heun, 'g', label='Heun Method')
plt.plot(t_rk, y_rk, 'b', label='Runge-Kutta Method')
plt.plot(t_exact, y_exact, 'k--', label='Exact Solution')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title('Solutions of the Differential Equation')
plt.show()
```
